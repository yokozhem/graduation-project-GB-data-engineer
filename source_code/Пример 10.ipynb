```python
import pandas as pd

# Загрузка данных
data = pd.read_csv('test.csv')

# Описание данных
description = data.describe()
print(description)
```

```python
import matplotlib.pyplot as plt

# Построение гистограммы
plt.hist(data['LotArea'], bins=30, edgecolor='k')
plt.title('Histogram of LotArea')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

# Построение диаграммы рассеяния
sns.scatterplot(x='LotArea', y='LotFrontage', data=data)
plt.title('Scatter Plot of LotArea vs LotFrontage')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()
```

```python
# Фильтрация только числовых столбцов
df_numeric = data.select_dtypes(include=['float64', 'int64'])

# Вычисление корреляционной матрицы
correlation_matrix = df_numeric.corr()

# Визуализация корреляционной матрицы
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
